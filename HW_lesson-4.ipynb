#### 1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800]. Найдите её среднее значение и дисперсию.

```python
M = (800 + 200) / 2
D = (800 - 200) ** 2 / 12

print(f'Среднее значение равномерно распределенной непрерывной случайной величины А равно {M}')
print(f'Дисперсия равномерно распределенной непрерывной случайной величины А = {D}')
```

#### 2. О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
#### Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
#### Если да, найдите ее.

По формуле дисперсии $D(X)$ равномерно распределенной непрерывной случайной величины можно вычислить правую границу величины В:
$$b = a + \sqrt{D(X) * 12}$$

```python
D = 0.2
A = 0.5
B = A + math.sqrt(D * 12)

print(f'Правая граница величины В равна {B:.4}')
```

```python
M = (A + B) / 2

print(f'Среднее значение величины В равно {B:.4}')
```

#### 3. Непрерывная случайная величина X распределена нормально и задана плотностью распределения
#### f(x) = (1 / (4 * sqrt(2*pi))) * (exp(-((x+2)**2) / 32))
#### Найдите:
#### а). M(X)
#### б). D(X)
#### в). std(X) (среднее квадратичное отклонение)

**Плотность вероятности нормального распределения** можно вычислить по формуле:$$f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e ^ {- \frac{(x - a)^2}{2 \sigma^2}},$$где $a = M(X), \; \sigma^2 = D(X).$

**Среднее квадратичное оклонение** можно вычислить по формуле: $$\sigma = \sqrt{\frac{\sum\limits_{i=1}^{n} (x_i - \overline{x})^2}{n}}$$

Значит,

а) M(X) = - 2

б) D(X) = 16

в) $\sigma$ (std(X)) = 4

#### 5. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

```python
d = (190 - 178) / math.sqrt(25)

print(f'Количество сигм отклонения роста человека от математического ожидания роста в популяции равно {d:.1f}')
```
